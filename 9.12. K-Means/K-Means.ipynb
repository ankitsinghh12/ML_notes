# K - Means Algorithm

## Clustering

- CLUSTERING TASK is to Group/ Cluster similar data points.

- **OBSERVATIONS** Points in a cluster are close together. Points in different cluster are far away from each other.

- Clustering is UNSUPERVISED LEARNING.

- Algorithm(for clustering): K-means, K-means++, Hierarchical Clustering, DBSCAN

## K-MEANS CLUSTERING 

Group them into 'k' clusters. For every clusters, it assigns centroid to each cluster.

## Applications of Clustering

1) E-commerce - Task is to group similar customers based on there purchasing behavior. Customer behaviour can include:
- How much money they spent
- Kind of credit card
- Kind of products they buy
- Geographical area



2) Image Segmentation - group similar pixel


3) After grouping, we typically apply ML techniques to perform object detection.

## Coding Sample

```python
import pandas as pd
import numpy as np
from sklearn import datasets

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
 
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

### main part 

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
print(model.labels_)
print(model.cluster_centers_)
```

```python
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
 
# Plot the Original Classifications
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[model.labels_])
plt.title('K Mean Classification')
```

```python
predict = model.predict(x)
```

```python
Y = iris.target
```

```python
from scipy.stats import mode
labels = np.zeros_like(predict)
for i in range(3):
    mask = (predict==i)
    labels[mask] = mode(Y[mask])[0]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(Y, labels)
```

```python
confusion_matrix(Y, labels)
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# Different ways to compute distance between two clusters
# Single Linkage

plt.figure(figsize=(12, 8))
mergings = linkage(x, method="single", metric="euclidean")
dendrogram(mergings)
plt.show()
```

```python
# Complete Linkage

plt.figure(figsize=(12, 8))
mergings = linkage(x, method="complete", metric="euclidean")
dendrogram(mergings)
plt.show()
```

### Three cuts we can consider using cut tree

```python
cut_tree(mergings, n_clusters=3)
```

```python
# 3 clusters

lables = cut_tree(mergings, n_clusters=3).reshape(-1, )

lables
```

```python
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
 
# Plot the Original Classifications
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[lables])
plt.title('Agglomerative Clustering')
```
